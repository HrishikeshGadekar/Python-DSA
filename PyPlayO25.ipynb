```python
def findArrayIntersection(arr1, arr2):
  freq_map = dict()
  intersection = []

  for i in arr1:
    freq_map[i] = freq_map.get(i, 0) + 1

  for i in arr2:
    if i in freq_map:
        intersection.append(i)
  
  return intersection

arr1 = [6, 9, 8, 5]
arr2 = [9, 2, 4, 1, 8]
findArrayIntersection(arr1, arr2)  # Expected Output: 9 8
```

```python
def compressString(s):
    result, count, n = [], 1, len(s)

    for (i, v) in enumerate(s):
        if (i < n-1) and (s[i] == s[i+1]):
            count += 1
            continue
        result.append(v)
        result.append(str(count))
        count = 1

    return "".join(result)

compressString("aabcccccaaa")
```

```python
def longestUniqueSubstring(s):
    char_map = {}
    start, maxLength = 0, 1

    for (i, v) in enumerate(s):
        if v in char_map and start <= char_map[v]:
            curr_max = i-start
            if curr_max > maxLength: maxLength = curr_max
            start = char_map[v]+1
        char_map[v] = i

    return maxLength

longestUniqueSubstring("abcade")  # Expected Output: "abc"
longestUniqueSubstring("aabcded")  # Expected Output: "abcde"
```

```python
def isValidPalindrome(s) -> bool:
    if not s: return False

    left, right = 0, len(s)-1

    while left < right:
        print(left, ":", s[left], right, ":", s[right])
        if s[left] == s[right]:
            left += 1
            right -= 1
            continue
        return False
    
    return True

isValidPalindrome("race a car")
```

```python
def findPattern(s: str, pattern: str) -> list:
    if not s: return False

    n, m = len(s), len(pattern)
    matches = []

    for i in range(n):
        if s[i: i+m] == pattern:
            matches.append(i)

    return matches

findPattern("abcabcab", "ab")
```

```python
class Solution:
    def containsNearbyDuplicate(self, nums: List[int], k: int) -> bool:
        n = len(nums)
        index_map = {}
        
        for (i, v) in enumerate(nums):
            if v in index_map and (i-index_map[v]) <= k: return True
            index_map[v] = i

        return False  
```

```python
def countDistinct(arr, k):
    if not arr: return []
    curr, n, output = 0, len(arr), []

    while curr <= n-k:
        count, matches = 0, set()
        for i in range(curr, curr+k):
            if arr[i] in matches: continue
            count += 1
            matches.add(arr[i])
        output.append(count)
        curr += 1

    return output

countDistinct([1, 2, 1, 3, 4, 2, 3], 4)
```

```python
def countDistinct(arr, k):
    if not arr: return []
    matches, output = {}, []

    for (i, v) in enumerate(arr):
        matches[v] = matches.get(v, 0) + 1
        if (i+1 >= k):
            output.append(len(matches))
            outgoing = arr[i+1-k]
            matches[outgoing] -= 1
            if matches[outgoing] == 0: del matches[outgoing]

    return output

countDistinct([1, 2, 1, 3, 4, 2, 3], 4)
```

```python
def longestConsecutive(nums) -> int:
    if not nums:return 0
    
    # Convert array to a set for O(1) lookups
    num_set = set(nums)
    max_length = 1

    for num in num_set:
        if num-1 not in num_set:
            curr, curr_length = num, 1
            while curr+1 in num_set:
                curr += 1
                curr_length += 1

            max_length = max(max_length, curr_length)

    return max_length
            

print(longestConsecutive([100, 4, 200, 1, 3, 2]))  # Expected: 4 (Sequence: [1, 2, 3, 4])
print(longestConsecutive([0, 3, 7, 2, 5, 8, 4, 6, 0, 1]))  # Expected: 9 (Sequence: [0, 1, 2, 3, 4, 5, 6, 7, 8])
print(longestConsecutive([]))  # Expected: 0 (Empty input)
print(longestConsecutive([1, 2, 0, 1]))  # Expected: 3 (Sequence: [0, 1, 2])
print(longestConsecutive([1]))  # Expected: 1 (Only one element)
```
