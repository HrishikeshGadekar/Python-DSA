### P1: Check Permutations

For a given two strings, 'str1' and 'str2', check whether they are a permutation of each other or not.

Permutations of each other
Two strings are said to be a permutation of each other when either of the string's characters can be rearranged so that it becomes identical to the other one.

Example:
str1= "sinrtg"
str2 = "string"

The character of the first string(str1) can be rearranged to form str2 and hence we can say that the given strings are a permutation of each other.

```python
def arePermutations(str1: str, str2: str) -> bool:
    if len(str1) != len(str2):
        return False  # Different lengths â†’ Not permutations

    freq = [0] * 256  # Frequency array for ASCII characters

    for ch in str1:
        freq[ord(ch)] += 1  # Increment count for str1

    for ch in str2:
        freq[ord(ch)] -= 1  # Decrement count for str2

    return (freq.count(0) == 256)   # If all counts are zero, they are permutations

# Example Usage
print(arePermutations("sinrtg", "string"))  # True
print(arePermutations("hello", "holle"))  # True
print(arePermutations("test", "ttew"))  # False
print(arePermutations("aabbcc", "ccbbaa"))  # True
print(arePermutations("abcd", "abce"))  # False
```

### P3: Highest Occuring Character

For a given a string(str), find and return the highest occurring character.

Example:
Input String: "abcdeapapqarr"
Expected Output: 'a'
Since 'a' has appeared four times in the string which happens to be the highest frequency character, the answer would be 'a'.
If there are two characters in the input string with the same frequency, return the character which comes first.

Consider:
Assume all the characters in the given string to be in lowercase always.

```python
def highestOccurringChar(string) :
    freq = [0]*256

    for i in string:
        freq[ord(i)] += 1

    max_count = max(freq)

    for i in string:
        if freq[ord(i)] == max_count:
            return i
    return ""

# Example Usage
print(highestOccurringChar("abcdeapapqarr"))  # Output: 'a'
print(highestOccurringChar("aaaaa"))          # Output: 'a'
print(highestOccurringChar("abbbbbccddddd"))  # Output: 'b' (comes first)
print(highestOccurringChar(""))               # Output: "" (empty string)
print(highestOccurringChar("a"))              # Output: 'a'
```

### P4: Compress the String

Write a program to do basic string compression. For a character which is consecutively repeated more than once, replace consecutive duplicate occurrences with the count of repetitions.

Example:
If a string has 'x' repeated 5 times, replace this "xxxxx" with "x5".
The string is compressed only when the repeated character count is more than 1.
Note:
Consecutive count of every character in the input string is less than or equal to 9. You are not required to print anything. It has already been taken care of. Just implement the given function and return the compressed string.

```python
def getCompressedString(s) :
  if not s: return ""
  result, prev, count = [], s[0], 0

  for curr in s:
    if prev == curr:
        count+= 1
        continue
    result.append(prev)
    if count > 1: result.append(str(count))
    prev, count = curr, 1

  result.append(prev)
  if count > 1: result.append(str(count))

  return "".join(result)

# Example Usage
print(getCompressedString("aaabbccdsa"))  # "a3b2c2dsa"
print(getCompressedString("aaabbcddeeeee"))  # "a3b2cd2e5"
print(getCompressedString("abc"))  # "abc" (No compression needed)
print(getCompressedString("a"))  # "a" (Single character)
print(getCompressedString(""))  # "" (Empty string)
```

### P10: Rotate Array (Reversal Algorithm)

Steps:
1. Write a reverse array method
2. Reverse elements from (0, d-1)
3. Reverse elements from (d, n-1)
4. Reverse the new array from (0, n-1)

```python
def rotateArray(arr, d):
  n = len(arr)
  if n == 0: return
  if d > n: d %= n

  def reverseArray(arr, s, e):
    for i in range((e - s + 1) // 2):
      arr[s + i], arr[e - i] = arr[e - i], arr[s + i]

  reverseArray(arr, 0, d-1);
  reverseArray(arr, d, n-1);
  reverseArray(arr, 0, n-1);

# Example Usage
arr = [1, 2, 3, 4, 5, 6, 7]
d = 2
rotateArray(arr, d)
print(arr)  # Expected Output: [3, 4, 5, 6, 7, 1, 2]
```

### P14: Find Duplicate Number

You have been given an integer array/list(ARR) of size N which contains numbers from 0 to (N - 2). Each number is present at least once. That is, if N = 5, the array/list constitutes values ranging from 0 to 3, and among these, there is a single integer value that is present twice. You need to find and return that duplicate number present in the array.

Note :
Duplicate number is always present in the given array/list.

```python
def findDuplicate(arr, n):
    xor_result = 0

    # XOR all elements in the array
    for num in arr:
        xor_result ^= num

    # XOR with all numbers from 0 to N-2
    for i in range(n-1):
        xor_result ^= i

    return xor_result

# Example usage
arr = [0, 1, 2, 3, 2]
print(findDuplicate(arr, 5))  # Expected Output: 2
```

### P16: Subarray Sum Equals K

https://leetcode.com/problems/subarray-sum-equals-k/description/?source=submission-noac

```python
def subarraySum(arr, k) -> int:
  curr_sum, count = 0, 0
  pre_sum = {0: 1}

  for num in arr:
    curr_sum += num
    count += pre_sum.get(curr_sum-k, 0)
    pre_sum[curr_sum] = pre_sum.get(curr_sum, 0) + 1

  return count

print(subarraySum([1,1,1], 2))  # Output: 2
print(subarraySum([1,2,3], 3))  # Output: 2 ([3], [1,2])
print(subarraySum([0,0,0,0], 0))  # Output: 10 (multiple overlapping subarrays)
print(subarraySum([1,-1, 0], 0))  # Output: 3
print(subarraySum([3,4,7,2,-3,1,4,2], 7))  # Output: 4
```

Find the longest substring without repeating characters.
<br/>Example: "abcabcbb" â†’ 3 ("abc")
<br/>https://leetcode.com/problems/longest-substring-without-repeating-characters/

```python
def lengthOfLongestSubstring(s) -> int:
    char_map = {}
    start, maxLength = 0, 0

    for (i, v) in enumerate(s):
        if v in char_map and start <= char_map[v]:
            start = char_map[v]+1
        char_map[v] = i
        maxLength = max(maxLength, i-start+1)

    return maxLength

lengthOfLongestSubstring("abcabcbb")
```

### P18: Count distinct elements within every subarray of length k

Approach:

*   
Use a sliding window with a frequency dictionary to maintain counts.
*   
Efficiently update the distinct count as the window slides.



```python
def countDistinct(arr, k):
    if not arr: return []
    matches, output = {}, []

    for (i, v) in enumerate(arr):
        matches[v] = matches.get(v, 0) + 1
        if (i+1 >= k):
            output.append(len(matches))
            outgoing = arr[i+1-k]
            matches[outgoing] -= 1
            if matches[outgoing] == 0: del matches[outgoing]

    return output

print(countDistinct([1, 2, 1, 3, 4, 2, 3], 4))  # Expected: [3, 4, 4, 3]
print(countDistinct([1, 2, 4, 4], 2))           # Expected: [2, 2, 1]
print(countDistinct([], 3))                     # Expected: []
print(countDistinct([1, 1, 1, 1], 1))           # Expected: [1, 1, 1, 1]
print(countDistinct([1, 2, 3, 4, 5], 6))        # Expected: []
```

### P19: Longest Consecutive Sequence

https://leetcode.com/problems/longest-consecutive-sequence/description/
image.png

```python
def longestConsecutive(nums) -> int:
    if not nums:return 0

    # Convert array to a set for O(1) lookups
    num_set = set(nums)
    max_length = 1

    for num in num_set:
        if num-1 not in num_set:
            curr, curr_length = num, 1
            while curr+1 in num_set:
                curr += 1
                curr_length += 1

            max_length = max(max_length, curr_length)

    return max_length

print(longestConsecutive([100, 4, 200, 1, 3, 2]))  # Expected: 4 (Sequence: [1, 2, 3, 4])
print(longestConsecutive([0, 3, 7, 2, 5, 8, 4, 6, 0, 1]))  # Expected: 9 (Sequence: [0, 1, 2, 3, 4, 5, 6, 7, 8])
print(longestConsecutive([]))  # Expected: 0 (Empty input)
print(longestConsecutive([1, 2, 0, 1]))  # Expected: 3 (Sequence: [0, 1, 2])
print(longestConsecutive([1]))  # Expected: 1 (Only one element)
```

### P21: Longest Subarray with Equal 0s and 1s

Given a binary array nums (containing only 0s and 1s), find the length of the longest subarray with an equal number of 0s and 1s.

```python
def longest_equal_zeros_ones(nums):
    sum_map = {0: -1}  # Store first occurrence of cumulative sum
    max_length, cum_sum = 0, 0

    for i, num in enumerate(nums):
        cum_sum += 1 if num == 1 else -1  # Convert 0 to -1

        if cum_sum in sum_map:
            max_length = max(max_length, i - sum_map[cum_sum])
        else:
            sum_map[cum_sum] = i  # Store first occurrence

    return max_length

print(longest_equal_zeros_ones([0, 1, 0, 1, 1, 0, 1]))
# Output: 6 (subarray: [0, 1, 0, 1, 1, 0])
```

### P24: Equillibrium Index

For a given array/list(ARR) of size 'N,' find and return the 'Equilibrium Index' of the array/list.

Equilibrium Index of an array/list is an index 'i' such that the sum of elements at indices [0 to (i - 1)] is equal to the sum of elements at indices [(i + 1) to (N-1)]. One thing to note here is, the item at the index 'i' is not included in either part.

If more than one equilibrium indices are present, then the index appearing first in left to right fashion should be returned. Negative one(-1) if no such index is present

```python
def arrayEquilibriumIndex(arr, n) :

    if n == 0 or n == 1: return -1
    leftSum, rightSum = 0, sum(arr)-arr[0]

    for i in range(1, n):
        leftSum += arr[i-1]
        rightSum -= arr[i]

        if leftSum == rightSum: return i

    return -1

arrayEquilibriumIndex([6, 1, 6, 5, 3, 2, 5, 0, 5, 6, 0], 11)
```

### P26: Find Unique Pairs with k difference

https://leetcode.com/problems/k-diff-pairs-in-an-array/

```python
def findPairs(nums, k) -> int:
    seen, pairs = set(), set()

    for (i, v) in enumerate(nums):
        target_1, target_2 = v-k, v+k
        if target_1 in seen and (v, target_1) not in pairs:
            pairs.add((target_1, v))
        if target_2 in seen and (v, target_2) not in pairs:
            pairs.add((target_2, v))
        seen.add(v)

    return len(pairs)

print(findPairs([3, 1, 4, 1, 5], 2))  # Expected Output: 2 (Pairs: (1,3) and (3,5))
print(findPairs([1, 2, 3, 4, 5], 1))  # Expected Output: 4 (Pairs: (1,2), (2,3), (3,4), (4,5))
print(findPairs([1, 3, 1, 5, 4], 0))  # Expected Output: 1 (Pair: (1,1))
print(findPairs([1, 2, 3, 4, 5], 10)) # Expected Output: 0 (No pairs)
print(findPairs([1, 2, 2, 3, 3], 1))  # Expected Output: 2 (Pairs: (1,2), (2,3))
```

```python
def mergeSort(head) :
  if head is None or head.next is None: return head

  def getMiddle(head):
      slow, fast = head, head.next
      while fast.next is not None and fast.next.next is not None:
          slow = slow.next
          fast = fast.next.next
      return slow

  def mergeSortedArrays(head1, head2):
      if head1 is None: return head2
      if head2 is None: return head1

      head, mergeHead = None, None
      if head1.data <= head2.data:
          head, mergeHead = head1, head1
          head1 = head1.next
      else:
          head, mergeHead = head2, head2
          head2 = head2.next

      while head1 is not None and head2 is not None:
          if head1.data <= head2.data:
              head.next = head1
              head1 = head1.next
          else:
              head.next = head2
              head2 = head2.next
          head = head.next

      head.next = head2 if head1 is None else head1

      return mergeHead

  # Step 1: Get the middle to split the list into two halves
  mid = getMiddle(head)
  # The second half starts at mid.next
  head1, head2 = head, mid.next
  mid.next = None # Breaks the list into two halves

  # Step 2: Recursively sort each half (DON'T WORRY ABOUT THIS SORT, IT'S RECURSIVE)
  sorted_head1 = mergeSort(head1)
  sorted_head2 = mergeSort(head2)

  # Step 3: Merge the two sorted halves
  return mergeSortedArrays(sorted_head1, sorted_head2)
```

### P34: Balanced Parantheses

```python
def isValid(expression) -> bool:
  openBrackets, closedBrackets = ('[', '{', '('), (']', '}', ')')
  brackets = {']': '[', '}': '{', ')':'('}

  stack = []
  for bracket in expression:
    if bracket in openBrackets:
      stack.append(bracket)

    if bracket in closedBrackets:
      if len(stack) == 0: return False
      if stack[-1] == brackets[bracket]: stack.pop()
      else: return False
  return not stack
```

#### Check Balanced Tree


**Solution 1:**

Time Complexity:
1. Best case: O(n*logn)
2. Worst case: O(n^2)

```python
class Solution:
  def isBalanced(self, root: Optional[TreeNode]) -> bool:
    if not root: return True

    leftSubTreeHeight = self.maxDepth(root.left)
    rightSubTreeHeight = self.maxDepth(root.right)

    if abs(leftSubTreeHeight - rightSubTreeHeight) > 1: return False

    return (self.isBalanced(root.left) and self.isBalanced(root.right))

  def maxDepth(self, root: Optional[TreeNode]) -> int:
    if not root: return 0
    return 1 + max(self.maxDepth(root.left), self.maxDepth(root.right))
```

**Solution 2:**

Time Complexity: O(n)

```python
def getHeightAndCheckBalanced(root):
  if not root: return 0, True

  leftSubTreeHeight, checkLeftBalance = getHeightAndCheckBalanced(root.left)
  rightSubTreeHeight, checkRightBalance = getHeightAndCheckBalanced(root.right)

  treeHeight = 1 + max(leftSubTreeHeight + rightSubTreeHeight)

  if abs(leftSubTreeHeight - rightSubTreeHeight) > 1: return treeHeight, False
  return treeHeight, (checkLeftBalance and checkRightBalance)
```

2. Using Post-order and In-order:

  ii. left, right Post-orders = preOrder[:inRoot], preOrder[inRoot:-1]

```python
class Solution:
    def buildTree(self, preOrder: List[int], inOrder: List[int]) -> Optional[TreeNode]:
        if len(preOrder) == 0: return

        rootData = preOrder[0]
        root = TreeNode(rootData)
        inRoot = inOrder.index(rootData)

        leftTreeInOrder, rightTreeInOrder = inOrder[:inRoot], inOrder[inRoot+1:]
        leftTreePreOrder, rightTreePreOrder = preOrder[1:inRoot+1], preOrder[inRoot+1:]

        leftTree = self.buildTree(leftTreePreOrder, leftTreeInOrder)
        rightTree = self.buildTree(rightTreePreOrder, rightTreeInOrder)

        root.left = leftTree
        root.right = rightTree

        return root
```

### P40: Hashmaps

```python
class MapNode:

  def __init__(self, key=None, value=None, next=None):
    self.key = key
    self.value = value
    self.next = next

class HashMap:

  def __init__(self):
    self.bucketSize = 10
    self.buckets = [None]*self.bucketSize
    self.count = 0

  def size(self):
    return self.count

  def getHashIndex(self, key):
    return (abs(hash(key))%(self.bucketSize))

  def getloadFactorAndCheckRehash(self):
    loadFactor = (self.count/self.bucketSize)
    return loadFactor, (loadFactor >= 0.7)

  def isNodePresent(self, head, key):
    if not head: return False
    while head:
      if head.key == key: return True
      head = head.next
    return False

  def getValue(self, key):
    if not key: return None
    index = self.getHashIndex(key)
    head = self.buckets[index]
    while head:
      if head.key == key: return head.value
      head = head.next
    return None

  def remove(self, key):
    if not key: return None
    index = self.getHashIndex(key)
    head = self.buckets[index]
    if not head: return None

    prev, curr = None, head
    while curr:
      if curr.key == key:
        if not prev: self.buckets[index] = curr.next
        else: prev.next = curr.next
        self.count -= 1
        return curr.value
      prev = curr
      curr = curr.next

    return None

  def reviseBuckets(self):
    self.bucketSize *= 2
    self.buckets = [None]*self.bucketSize
    self.count = 0

  def rehash(self):
    currentBuckets = self.buckets
    self.reviseBuckets()
    for currHead in currentBuckets:
      while currHead:
        self.insert(currHead.key, currHead.value)
        currHead = currHead.next

  def insert(self, key, value):
    self.count += 1
    index = self.getHashIndex(key)
    head = self.buckets[index]
    if not self.isNodePresent(head, key):
      newNode = MapNode(key, value)
      newNode.next = head
      self.buckets[index] = newNode
    else:
      while head:
        if head.key == key:
          head.value = value
          return
        head = head.next
    if self.getloadFactorAndCheckRehash()[1]:
      self.rehash()



m = HashMap()
m.insert('Parikh', 2)
print(m.size())
m.insert("Rohan", 7)
print(m.size())
m.insert('Parikh',9)
print(m.size())
print(m.getValue('Rohan'))
print(m.remove('Rohan'))
print(m.getValue('Rohan'))


print("-----Rehash Map------")
m1 = HashMap()
for i in range(10):
  m1.insert('abc' + str(i), i+1)
  print(m1.getloadFactorAndCheckRehash()[0])
```
